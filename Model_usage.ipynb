```python
from tensorflow import keras
model = keras.models.load_model(r'C:\OpenCV\Spot_Quality_Detection\Final CNN_2\Rotten\Fresh Fruits.h5')
```

```python
import cv2
from keras.preprocessing import image
import tensorflow as tf
import numpy as np
import os

img_path=r'C:\OpenCV\Spot_Quality_Detection\Final CNN\CNN_2\rotated_by_15_Screen Shot 2018-06-12 at 11.18.46 PM.png'
img=cv2.imread(img_path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img=cv2.resize(img,(224,224))
test_image = tf.keras.preprocessing.image.img_to_array(img)
test_image = np.expand_dims(test_image, axis = 0)

print(test_image.shape)
# cv2.imshow('Test',img)
# cv2.waitKey()

Predictions=model.predict(test_image)
Max_val=np.argmax(Predictions)
old_dict={0: 'fresh_apples', 1: 'fresh_banana', 2: 'fresh_oranges', 3: 'rotten_apples', 4: 'rotten_banana', 5: 'rotten_oranges'}
print('Old _Classes :',old_dict)
 
# new_dict = dict([(value, key) for key, value in old_dict.items()])
# print(new_dict)
print('Max_val: ',Max_val)
print('Predicted Fruit :',old_dict[Max_val])

# if Max_val==0:
#     print("Fresh Apple")
# if Max_val==1:
#     print("Fresh Banana")
# if Max_val==2:
#     print("Fresh Orange")
# if Max_val==3:
#     print('Rotten Apple')
# if Max_val==4:
#     print('Rotten Banana')
# if Max_val==5:
#     print('Rotten Orange')

```

```python
vid = cv2.VideoCapture(0)
print("Camera connection successfully established")
i = 0
while(True):  
    r, frame = vid.read() 
    cv2.imshow('frame', frame)
    cv2.imwrite(r'C:\OpenCV\Spot_Quality_Detection\Final CNN\CNN_2\Cam_Images'+str(i)+".jpg", frame)
    test_image = tf.keras.preprocessing.image.load_img(r'C:\OpenCV\Spot_Quality_Detection\Final CNN\CNN_2\Cam_Images'+str(i)+".jpg", target_size = (224, 224))
    test_image = tf.keras.preprocessing.image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)

    Predictions=model.predict(test_image)
    Max_val=np.argmax(Predictions)
    print('Old _Classes :',old_dict)
 
    # new_dict = dict([(value, key) for key, value in old_dict.items()])
    # print(new_dict)
    print('Max_val: ',Max_val)
    print('Predicted Fruit :',old_dict[Max_val])

    cv2.putText(frame, old_dict[Max_val], (50,50), cv2.FONT_HERSHEY_SIMPLEX, 
                   1, (255,0,0), 2, cv2.LINE_AA)
    cv2.imshow('Predicted Image',frame)
    os.remove(r'C:\OpenCV\Spot_Quality_Detection\Final CNN\CNN_2\Cam_Images'+str(i)+".jpg")
    i = i + 1
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
vid.release() 
cv2.destroyAllWindows() 
```
